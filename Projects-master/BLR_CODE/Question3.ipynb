```python

def parse_line(line):
    match = re.match(r'^(\d+)([A-Za-z\' ]+)(GS|GO|GP)', line)  
    if match:
        wearer_number = match.group(1)
        wearer_name = match.group(2).strip()
        wearer_first, wearer_last = wearer_name.rsplit(' ', 1) 
        item_code = match.group(3)  # Extract item code
        item_info = re.findall(r'(GS|GO|GP)(\d{4})(.*?)\s*(\d+\.\d+)', line)
        output = []
        for item in item_info:
            item_description = re.findall(r'SHIRT WORK LS|SHIRT WORK SS|COVERALL GAS CO COTTON|PANT WORK TWILL', item[2])[0]
            cost = item[3]
            output.append([wearer_number, wearer_first, wearer_last, item_code + item[1], item_description, cost])
        return output
    else:
        return None

data = []
for line in lines:
    output = parse_line(line)
    if output:
        data.extend(output)

df = pd.DataFrame(data, columns=['wearer_number', 'wearer_first', 'wearer_last', 'item_code', 'item_description', 'cost'])

df
```
