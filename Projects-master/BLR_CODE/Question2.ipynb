```python
url = "https://www.tfrrs.org/all_performances/IN_college_m_Indiana_IN.html?list_hnd=4153&season_hnd=608#top"
response = requests.get(url) ## to get access to the html code.

soup = BeautifulSoup(response.content, "html.parser") ## read in the html


table = soup.find("table", {"id": "myTable"}) ## Finding our information(Table we want)

data = []  ## Building in the table, this reads in one line
for row in table.find_all("tr", {"class": "allRows"}):
    cells = row.find_all("td")
    athlete = cells[1].text.strip()
    year = cells[2].text.strip()
    time = cells[3].text.strip()
    meet = cells[4].text.strip()
    meet_date = cells[5].text.strip()
    wind = cells[6].text.strip()
    data.append((athlete, year, time, meet, meet_date, wind)) ## Puts variable back into the table.

## Right here we build the table.
Indiana_df = pd.DataFrame(data, columns=["Athlete", "Year", "Time", "Meet", "Meet Date", "Wind"]) 
Indiana_df.head()
```

```python
athlete, year, time, meet, meet_date, wind
```

```python
# Split the string up by the "," using regex, and then flip flop them. 
# (Last name, first name ) --> (first name, Last name)
Indiana_df[['Last Name', 'First Name']] = Indiana_df['Athlete'].str.split(', ', expand=True)
# We flip it around.
Indiana_df['Athlete'] = Indiana_df['First Name'] + ' ' + Indiana_df['Last Name']
Indiana_df.drop(columns=['First Name', 'Last Name'], inplace=True)
Indiana_df
```

```python
## Regex to get just get the number for the year. 
Indiana_df['Year'] = Indiana_df['Year'].apply(lambda x: re.sub(r'\D', '', x))
## to look at it if done correctly.
Indiana_df
```

```python
## We need to change the format first, and then we do epoch
Indiana_df['Meet Date'] = pd.to_datetime(Indiana_df['Meet Date'], format='%b %d, %Y')

# using the epoch time
Indiana_df['Meet Date'] = Indiana_df['Meet Date'].apply(lambda x: int(x.timestamp()))

## see if its done correclt.y
Indiana_df
```

```python
# We want the fastest, so we group, get the min, the loweset number thet the mile done.
Indiana_fastest_times = Indiana_df.groupby('Athlete')['Time'].min()

## see the result (this isn't dataframe)
Indiana_fastest_times
```

```python
## Just resseting it to dataframe.
Indiana_fastest_times_df = Indiana_fastest_times.reset_index()
Indiana_fastest_times_df
```

```python

url = "https://www.tfrrs.org/lists/4768/Pac_12_Outdoor_Performance_List"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
table = soup.find("table", {"id": "myTable"})

data = []
for row in table.find_all("tr", {"class": "allRows"}):
    # Extracting individual cells from each row
    cells = row.find_all("td")
    # Extracting data from each cell
    rank = cells[0].text.strip()
    athlete = cells[1].text.strip()
    year = cells[2].text.strip()
    mark = cells[3].text.strip()
    meet = cells[4].text.strip()
    meet_date = cells[5].text.strip()
    data.append((rank, athlete, year, mark, meet, meet_date))
Outdoor_Performance = pd.DataFrame(data, columns=["Rank", "Athlete", "Year", "Mark", "Meet", "Meet Date"])
## Our data name
Outdoor_Performance
```

```python
# Split the string up by the "," using regex, and then flip flop them. 
# (Last name, first name ) --> (first name, Last name)
Outdoor_Performance[['Last Name', 'First Name']] = Outdoor_Performance['Athlete'].str.split(', ', expand=True)
Outdoor_Performance['Athlete'] = Outdoor_Performance['First Name'] + ' ' + Outdoor_Performance['Last Name']
Outdoor_Performance.drop(columns=['First Name', 'Last Name'], inplace=True)

Outdoor_Performance
```

```python
# Drop the letters, and only keep years.
Outdoor_Performance['Year'] = Outdoor_Performance['Year'].apply(lambda x: re.sub(r'\D', '', x))
Outdoor_Performance
```

```python
# This an extra step I took to fix the data in the Meet column
Outdoor_Performance['Meet'] = Outdoor_Performance['Meet'].str.extract(r'(\d+\.\d+)').astype(float)
Outdoor_Performance
```

```python
# This was to check if there are duplicates in there. (could of done unique as well)
fastest_meet = Outdoor_Performance.groupby('Athlete')['Meet'].max()

fastest_Time = pd.merge(Outdoor_Performance, fastest_meet, on=['Athlete', 'Meet'])
fastest_Time
```

```python
## renaming the column to help merge easily.
Outdoor_Performance = fastest_Time[['Athlete', 'Meet']].rename(columns={'Meet': 'Time'})
Outdoor_Performance
```
